```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2

inputfilepath = 'C:\EMOLGUI\A_O_C\Day12.txt'
```

```python
Map = []
with open(inputfilepath,'r') as f:
    for line in f:
        line = list(line.rstrip())
        if 'S' in line:
            s_col = line.index('S')
            s_row = len(Map)
            line[s_col] = 'a'
            startPosition = (s_row, s_col)
        if 'E' in line:
            e_col = line.index('E')
            e_row = len(Map)
            line[e_col] = 'z'
            endPosition = (e_row, e_col)
        Map.append(line)
numRows = len(Map)
numCols = len(Map[0])
```

```python
print(startPosition, endPosition, numRows, numCols)
print(Map)
```

```python
def testCandidatePosition(Map, candidatePosition, currentHeight, endPosition, visitedPositions, pathList, currentPath):
    # Already visited in another (shorter) path
    if candidatePosition in visitedPositions:
        return
    candidateHeight = Map[candidatePosition[0]][candidatePosition[1]]
    # Not reachable from current position
    if ord(candidateHeight) - ord(currentHeight) > 1:
        return
    visitedPositions.add(candidatePosition)
    nextPath = currentPath.copy()
    nextPath.append(candidatePosition)
    #print(nextPath)
    pathList.append(nextPath)
    return
```

```python
finished = False
pathList = [[startPosition]]
visitedPositions = set([startPosition])
while pathList:
    path = pathList.pop(0)
    currentRow, currentCol = path[-1][0], path[-1][1]
    currentHeight = Map[currentRow][currentCol]
    if currentRow > 0:
        nextCandidatePosition = (currentRow - 1, currentCol)
        testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path)
    if currentRow < numRows - 1:
        nextCandidatePosition = (currentRow + 1, currentCol)
        testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path)
    if currentCol > 0:
        nextCandidatePosition = (currentRow , currentCol - 1)
        testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path)
    if currentCol < numCols - 1:
        nextCandidatePosition = (currentRow , currentCol + 1)
        testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path)
    if endPosition in visitedPositions:
        break
while True:
    bestPath = pathList.pop()
    if endPosition in bestPath:
        break
```

```python
def testCandidatePosition(Map, candidatePosition, currentHeight, endPosition, visitedPositions, pathList, currentPath):
    # Already visited in another (shorter) path
    if candidatePosition in visitedPositions:
        return currentHeight
    candidateHeight = Map[candidatePosition[0]][candidatePosition[1]]
    # Not reachable from current position
    if ord(currentHeight) - ord(candidateHeight) > 1:
        return currentHeight
    visitedPositions.add(candidatePosition)
    nextPath = currentPath.copy()
    nextPath.append(candidatePosition)
    #print(nextPath)
    pathList.append(nextPath)
    return candidateHeight
```

```python
lowestHeight = 'z'
pathList = [[endPosition]]
visitedPositions = set([startPosition])
while pathList:
    path = pathList.pop(0)
    currentRow, currentCol = path[-1][0], path[-1][1]
    currentHeight = Map[currentRow][currentCol]
    if currentRow > 0:
        nextCandidatePosition = (currentRow - 1, currentCol)
        lowestHeight = min(lowestHeight, testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path))
    if currentRow < numRows - 1:
        nextCandidatePosition = (currentRow + 1, currentCol)
        lowestHeight = min(lowestHeight, testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path))
    if currentCol > 0:
        nextCandidatePosition = (currentRow , currentCol - 1)
        lowestHeight = min(lowestHeight, testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path))
    if currentCol < numCols - 1:
        nextCandidatePosition = (currentRow , currentCol + 1)
        lowestHeight = min(lowestHeight, testCandidatePosition(Map, nextCandidatePosition, currentHeight, endPosition, visitedPositions, pathList, path))
    if lowestHeight == 'a':
        break
while True:
    bestPath = pathList.pop()
    if Map[bestPath[-1][0]][bestPath[-1][1]] == 'a':
        break
```
