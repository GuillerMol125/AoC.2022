```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy
import logging

inputfilepath = 'C:\EMOLGUI\A_O_C\Day22.txt'
#inputfilepath = 'C:\EMOLGUI\A_O_C\Day22.sample.txt'
```

```python
board = []
with open(inputfilepath,'r') as f:
    for line in f:
        row = list(line.rstrip())
        
        board.append(row)
instructions = board.pop(-1)
board.pop(-1)
```

```python
directionMap = {'right':{'R':'down', 'L':'up', 'Value':0},
                'left':{'R':'up', 'L':'down', 'Value':2},
                'up':{'R': 'right', 'L': 'left', 'Value':3},
                'down':{'R': 'left', 'L': 'right', 'Value':1}}
```

```python
instLine = ''.join(instructions)
inst_splitR = instLine.split('R')
finalInst = []
for miniInst in inst_splitR:
    for item in miniInst.split('L'):
        finalInst.append(item)
        finalInst.append('L')
    finalInst.pop(-1)
    finalInst.append('R')
finalInst.pop(-1)
finalInst, len(finalInst)
```

```python
np_board = np.zeros([len(board),len(board[0])])
np_board.shape
for rowI, row in enumerate(board):
    for colI, col in enumerate(row):
        if col == '.':
            np_board[rowI,colI] = 1
        if col == '#':
            np_board[rowI,colI] = 2
np_board    
```

```python
starting_position = [0, np.min(np.where(np_board[0,:] == 1))]
starting_direction = 'right'
starting_position, starting_direction
```

```python
def move1(np_board, position, direction):
    next_position = [position[0], position[1]]
    if direction == 'right':
        next_position = [next_position[0], (next_position[1] + 1) % np_board.shape[1]]
        while np_board[next_position[0],next_position[1]] == 0:
            next_position = [next_position[0], (next_position[1] + 1) % np_board.shape[1]]
        if np_board[next_position[0],next_position[1]] == 1:
            return next_position
        if np_board[next_position[0],next_position[1]] == 2:
            return position

    if direction == 'left':
        next_position = [next_position[0], (next_position[1] - 1) % np_board.shape[1]]
        while np_board[next_position[0],next_position[1]] == 0:
            next_position = [next_position[0], (next_position[1] - 1) % np_board.shape[1]]
        if np_board[next_position[0],next_position[1]] == 1:
            return next_position
        if np_board[next_position[0],next_position[1]] == 2:
            return position

    if direction == 'up':
        next_position = [(next_position[0] - 1) % np_board.shape[0], next_position[1] ]
        while np_board[next_position[0],next_position[1]] == 0:
            next_position = [(next_position[0] - 1) % np_board.shape[0], next_position[1] ]
        if np_board[next_position[0],next_position[1]] == 1:
            return next_position
        if np_board[next_position[0],next_position[1]] == 2:
            return position

    if direction == 'down':
        next_position = [(next_position[0] + 1) % np_board.shape[0], next_position[1] ]
        while np_board[next_position[0],next_position[1]] == 0:
            next_position = [(next_position[0] + 1) % np_board.shape[0], next_position[1] ]
        if np_board[next_position[0],next_position[1]] == 1:
            return next_position
        if np_board[next_position[0],next_position[1]] == 2:
            return position    

def moveN(np_board, position, direction, N):
    next_position = [position[0], position[1]]
    for repeatI in range(N):
        #print(f'    Step {repeatI}')
        next_position = move1(np_board, next_position, direction)      
    return next_position 
```

```python
position = [starting_position[0], starting_position[1]]
direction = starting_direction
instList = copy.deepcopy(finalInst)
while instList:
    nextN = int(instList.pop(0))
    print(f'Moving {direction} : {nextN} from {position}')
    position = moveN(np_board, position, direction, nextN)
    if len(instList) == 0:
        break
    #if len(finalInst) % 25 == 0:
    print(f'Remaining instructions: {len(instList)}')
    nextRot = instList.pop(0)
    direction = directionMap[direction][nextRot]
position, direction, 1000*(position[0]+1) + 4*(position[1]+1) + directionMap[direction]['Value']
```

```python
Face1 = {'range_row':[0, 49],
'range_col':[50, 99],
'right':['Face2', 'left', 'direct'], 'left':['Face5', 'left', 'inverse'], 'up':['Face4', 'left', 'direct'], 'down':['Face3', 'up', 'direct']}
Face2 = {'range_row':[0, 49],
'range_col':[100, 149],
'right':['Face6', 'right', 'inverse'], 'left':['Face1', 'right', 'direct'], 'up':['Face4', 'down', 'direct'], 'down':['Face3', 'right', 'direct']}
Face3 = {'range_row':[50, 99],
'range_col':[50, 99],
'right':['Face2', 'down', 'direct'], 'left':['Face5', 'up', 'direct'], 'up':['Face1', 'down', 'direct'], 'down':['Face6', 'up', 'direct']}
Face5 = {'range_row':[100, 149],
'range_col':[0, 49],
'right':['Face6', 'left', 'direct'], 'left':['Face1', 'left', 'inverse'], 'up':['Face3', 'left', 'direct'], 'down':['Face4', 'up', 'direct']}
Face6 = {'range_row':[100, 149],
'range_col':[50, 99],
'right':['Face2', 'right', 'inverse'], 'left':['Face5', 'right', 'direct'], 'up':['Face3', 'down', 'direct'], 'down':['Face4', 'right', 'direct']}
Face4 = {'range_row':[150, 199],
'range_col':[0, 49],
'right':['Face6', 'down', 'direct'], 'left':['Face1', 'up', 'direct'], 'up':['Face5', 'down', 'direct'], 'down':['Face2', 'up', 'direct']}
Cube = {}
Cube['Face1'] = Face1
Cube['Face2'] = Face2
Cube['Face3'] = Face3
Cube['Face4'] = Face4
Cube['Face5'] = Face5
Cube['Face6'] = Face6
```

```python
starting_position = [0, np.min(np.where(np_board[0,:] == 1))]
starting_direction = 'right'
starting_face = 'Face1'
starting_position, starting_direction, starting_face
```

```python
def changeFace(np_board, face, origin, margin, mode, cube):
    next_face = cube[face]
    if origin == 'left':
        next_direction = 'right'
        next_col = min(next_face['range_col'])
        if mode == 'inverse':
            next_row = max(next_face['range_row']) - margin
        elif mode == 'direct':
            next_row = min(next_face['range_row']) + margin
    if origin == 'right':
        next_direction = 'left'
        next_col = max(next_face['range_col'])
        if mode == 'inverse':
            next_row = max(next_face['range_row']) - margin
        elif mode == 'direct':
            next_row = min(next_face['range_row']) + margin
    if origin == 'up':
        next_direction = 'down'
        next_row = min(next_face['range_row'])
        if mode == 'inverse':
            next_col = max(next_face['range_col']) - margin
        elif mode == 'direct':
            next_col = min(next_face['range_col']) + margin
    if origin == 'down':
        next_direction = 'up'
        next_row = max(next_face['range_row'])
        if mode == 'inverse':
            next_col = max(next_face['range_col']) - margin
        elif mode == 'direct':
            next_col = min(next_face['range_col']) + margin
    next_position = [next_row, next_col]
    print(f'    Switching to face {face} entering from {origin} in mode {mode} with margin {margin}')
    return next_direction, next_position


def move1(np_board, position, direction, face, cube):
    next_position = [position[0], position[1]]
    next_direction = direction
    this_face = cube[face]
    next_face = face

    if next_direction == 'right':
        next_position = [next_position[0], (next_position[1] + 1)]

        if next_position[1] > max(this_face['range_col']):
            next_face, origin, mode = this_face[next_direction]
            margin = next_position[0] - min(this_face['range_row'])
            next_direction, next_position = changeFace(np_board, next_face, origin, margin, mode, cube)

        if np_board[next_position[0],next_position[1]] == 1:
            return next_position, next_direction, next_face
        if np_board[next_position[0],next_position[1]] == 2:
            return position, direction, face

    if next_direction == 'left':
        next_position = [next_position[0], (next_position[1] - 1)]

        if next_position[1] < min(this_face['range_col']):
            next_face, origin, mode = this_face[next_direction]
            margin = next_position[0] - min(this_face['range_row'])
            next_direction, next_position = changeFace(np_board, next_face, origin, margin, mode, cube)

        if np_board[next_position[0],next_position[1]] == 1:
            return next_position, next_direction, next_face
        if np_board[next_position[0],next_position[1]] == 2:
            return position, direction, face

    if next_direction == 'up':
        next_position = [(next_position[0] - 1), next_position[1] ]
        if next_position[0] < min(this_face['range_row']):
            next_face, origin, mode = this_face[next_direction]
            margin = next_position[1] - min(this_face['range_col'])
            next_direction, next_position = changeFace(np_board, next_face, origin, margin, mode, cube)

        if np_board[next_position[0],next_position[1]] == 1:
            return next_position, next_direction, next_face
        if np_board[next_position[0],next_position[1]] == 2:
            return position, direction, face

    if next_direction == 'down':
        next_position = [(next_position[0] + 1), next_position[1] ]
        if next_position[0] > max(this_face['range_row']):
            next_face, origin, mode = this_face[next_direction]
            margin = next_position[1] - min(this_face['range_col'])
            next_direction, next_position = changeFace(np_board, next_face, origin, margin, mode, cube)

        if np_board[next_position[0],next_position[1]] == 1:
            return next_position, next_direction, next_face
        if np_board[next_position[0],next_position[1]] == 2:
            return position, direction, face  

def moveN(np_board, position, direction, face, N, cube):
    next_position = [position[0], position[1]]
    next_direction = direction
    next_face = face
    for repeatI in range(N):
        #print(f'    Step {repeatI}')
        next_position, next_direction, next_face  = move1(np_board, next_position, next_direction, next_face, cube)  

    return next_position, next_direction, next_face
```
