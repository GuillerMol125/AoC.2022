```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy
import logging

inputfilepath = 'C:\EMOLGUI\A_O_C\Day25.txt'
#inputfilepath = 'C:\EMOLGUI\A_O_C\Day25.sample.txt'
```

```python
totalValue = 0
with open(inputfilepath,'r') as f:
    for line in f:
        rowItems = list(line.rstrip())
        power = 1
        result = 0
        while rowItems:
            item = rowItems.pop()
            if item == '-':
                value = -1
            elif item == '=':
                value = -2
            else:
                value = int(item)
            result += value*power
            power = power*5        
        print(f'{line} corresponds to {result}')
        totalValue += result
            
print(f'Total value is {totalValue}')
```

```python
# Convert to base 5
totalValueBase5 = []
remainingValue = totalValue
power = 1
while remainingValue >= 5*power:
    power = power*5
while power >= 1:
    digit = int(remainingValue/power)
    totalValueBase5.append(digit)
    remainingValue -= digit*power
    power = power/5
print(f'Value {totalValue} in base 5 is {totalValueBase5}')
```

```python
# Convert to SNAFU 2,1,0,-,=
totalValueSnafu = copy.copy(totalValueBase5)
for i in range(len(totalValueSnafu)-1,0,-1):
    digit = totalValueSnafu[i]
    if digit <= 2:
        totalValueSnafu[i] = str(digit)
    elif digit == 3:
        totalValueSnafu[i-1] += 1
        totalValueSnafu[i] = '='
    elif digit == 4:
        totalValueSnafu[i-1] += 1
        totalValueSnafu[i] = '-'
    elif digit == 5:
        totalValueSnafu[i-1] += 1
        totalValueSnafu[i] = '0'
    else:
        print(f'Error detected, digit too high: {digit}')
        break

digit = totalValueSnafu[0]
if digit <= 2:
    totalValueSnafu[0] = str(digit)
elif digit == 3:
    totalValueSnafu[0] = '='
    totalValueSnafu = [1] + totalValueSnafu
    
elif digit == 4:
    totalValueSnafu[0] = '-'
    totalValueSnafu = [1] + totalValueSnafu
    
elif digit == 5:
    totalValueSnafu[0] = '0'
    totalValueSnafu = [1] + totalValueSnafu
    
```
