```python
import os
import sys
import numpy as np

inputfilepath = 'C:\EMOLGUI\A_O_C\Day4.txt'
```

```python
with open(inputfilepath,'r') as f:
    assignments_list = []
    full_contain = 0
    for line in f:
        assignments = line.rstrip().rsplit(',')
        firstPart = assignments[0].rsplit('-')
        secondPart = assignments[1].rsplit('-')
        check = (int(firstPart[0]) - int(secondPart[0]))*(int(firstPart[1]) - int(secondPart[1]))
        if check <= 0:
            contains = 1
        else:
            contains = 0
        assignments_list.append([firstPart,secondPart,contains])
```

```python
assignments_list
```

```python
np_assignments_list = np.array(assignments_list)
len(np_assignments_list), np.sum(np_assignments_list[:,2])
```

#### QUESTION 2: HOW MANY PARTS OVERLAP?

```python
with open(inputfilepath,'r') as f:
    assignments_list = []
    for line in f:
        assignments = line.rstrip().rsplit(',')
        firstPart = assignments[0].rsplit('-')
        secondPart = assignments[1].rsplit('-')
        overlap = max(min(int(firstPart[1]),int(secondPart[1])) - max(int(firstPart[0]),int(secondPart[0])) + 1,0)
        assignments_list.append([firstPart,secondPart,overlap,overlap>0])
```

```python
np_assignments_list = np.array(assignments_list)
assignments_list
```

```python
len(np_assignments_list), np.sum(np_assignments_list[:,3])
```
