```python
import os
import sys
import numpy as np

inputfilepath = 'C:\EMOLGUI\A_O_C\Day2.txt'
```

```python
rockPaperScissors = {'A':{'X':3, 'Y':6, 'Z':0}, 'B':{'X':0, 'Y':3, 'Z':6}, 'C':{'X':6, 'Y':0, 'Z':3}}
symbolValue = {'X':1, 'Y':2, 'Z':3}
```

```python
with open(inputfilepath,'r') as f:
    round_result_list = []
    round_list = []
    for line in f:
        round = line.rstrip().rsplit(' ')
        round_result = [symbolValue[round[1]], rockPaperScissors[round[0]][round[1]]]
        round_result_list.append(round_result)
        round_list.append(round + round_result)
```

### QUESTION 1: WHAT WOULD THE TOTAL SCORE BE?

```python
len(round_result_list), np.sum(round_result_list)
```
