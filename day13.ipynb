```python
def comparePackets(left,right,ordering):
    if type(left) == int and type(right) == int:
        if left < right:
            ordering.append(1)
            return True
        if left > right:
            ordering.append(-1)
            return True
        else:
            return False

    if type(left) == int and type(right) != int:
        return comparePackets([left],right,ordering)
    if type(left) != int and type(right) == int:
        return comparePackets(left,[right],ordering)

    if len(left) == 0 and len(right) > 0:
        ordering.append(1)
        return True
    if len(left) > 0 and len(right) == 0:
        ordering.append(-1)
        return True
    if len(left) == 0 and len(right) == 0:
        return False
    
    left_first = left.pop(0)
    right_first = right.pop(0)
    if comparePackets(left_first,right_first,ordering):
        return True
    else:
        return comparePackets(left,right,ordering)
```

```python
for index in range(len(buffer)-1):
    print(f'Getting position {index}')
    for compIndex in range(index+1,len(buffer)):
        compRes = []
        left = copy.deepcopy(buffer[index])
        right = copy.deepcopy(buffer[compIndex])
        comparePackets(left,right,compRes)
        if compRes[-1] == -1:
            buffer[index], buffer[compIndex] = buffer[compIndex], buffer[index]        
```
