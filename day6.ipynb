#### QUESTION 2: 14 different chars instead of just 4

```python
with open(inputfilepath,'r') as f:
    line = f.readline().rstrip()
    chars = list(line)
    charIndex = 14
    previousChars = set(chars[charIndex-14:charIndex])
    while len(previousChars) < 14:
        print(chars[charIndex-14:charIndex],previousChars,len(previousChars),charIndex)
        charIndex += 1
        previousChars = set(chars[charIndex-14:charIndex])
```
