```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy
import logging

inputfilepath = 'C:\EMOLGUI\A_O_C\Day23.txt'
#inputfilepath = 'C:\EMOLGUI\A_O_C\Day23.sample.txt'
```

```python
def rangeElves(elves):
    minX = elves[0][0]
    maxX = elves[0][0]
    minY = elves[0][1]
    maxY = elves[0][1]
    for elf in elves:
        minX = min(minX, elf[0])
        maxX = max(maxX, elf[0])
        minY = min(minY, elf[1])
        maxY = max(maxY, elf[1])
    return minX, maxX, minY, maxY

def drawElves(elves):
    minX, maxX, minY, maxY = rangeElves(elves)
    elveMatrix = []
    for x in range(minX, maxX+1):
        line = []
        for y in range(minY, maxY+1):
            if (x, y) in elves:
                line.append('#')
            else:
                line.append('.')
        print(''.join(line))
        elveMatrix.append(line)
    return elveMatrix
```

```python
elves = []
rowCount = 1
with open(inputfilepath,'r') as f:
    for line in f:
        colCount = 1
        row = list(line.rstrip())
        for item in row:
            if item == '#':
                elf = (rowCount, colCount)
                elves.append(elf)
            colCount += 1
        rowCount += 1
```

```python
moveOrder = ['north', 'south', 'west', 'east']
maxRounds = 10

for round in range(maxRounds):

    print(f'Round number {round+1}')

    nextMoveList = []
    repeatedMoves = set()

    for elf in elves:
        Npos = (elf[0]-1, elf[1])
        NEpos = (elf[0]-1, elf[1]+1)
        NWpos = (elf[0]-1, elf[1]-1)
        Spos  = (elf[0]+1, elf[1])
        SEpos  = (elf[0]+1, elf[1]+1)
        SWpos  = (elf[0]+1, elf[1]-1)
        Wpos  = (elf[0], elf[1]-1)
        Epos  = (elf[0], elf[1]+1)
        if not (Npos in elves or NEpos in elves or NWpos in elves or Spos in elves or SEpos in elves or SWpos in elves or Wpos in elves or Epos in elves):
            # do not move
            nextMoveList.append(elf)
            continue
        for nextMove in moveOrder:
            if nextMove == 'north':
                if not (Npos in elves or NEpos in elves or NWpos in elves):
                    if (Npos) in nextMoveList:
                        repeatedMoves.add(Npos)
                    nextMoveList.append(Npos)
                    break

            if nextMove == 'south':
                if not (Spos in elves or SEpos in elves or SWpos in elves):
                    if (Spos) in nextMoveList:
                        repeatedMoves.add(Spos)
                    nextMoveList.append(Spos)
                    break

            if nextMove == 'west':
                if not (Wpos in elves or NWpos in elves or SWpos in elves):
                    if (Wpos) in nextMoveList:
                        repeatedMoves.add(Wpos)
                    nextMoveList.append(Wpos)
                    break

            if nextMove == 'east':
                if not (Epos in elves or NEpos in elves or SEpos in elves):
                    if (Epos) in nextMoveList:
                        repeatedMoves.add(Epos)
                    nextMoveList.append(Epos)
                    break
        else:
            nextMoveList.append(elf)

    for elfIndex, nextMove in enumerate(nextMoveList):
        if not (nextMove in repeatedMoves):
            elves[elfIndex] = nextMove
    
    moveOrder.append(moveOrder.pop(0))
    
    drawElves(elves)
    print('')
    print('')
    print('')
```
