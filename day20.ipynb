```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy
import logging

inputfilepath = 'C:\EMOLGUI\A_O_C\Day20.txt'
#inputfilepath = 'C:\EMOLGUI\A_O_C\Day20.sample.txt'
```

```python
class Item():
    def __init__(self, value, order):
        self.value = value
        self.chronological_order = order
        self.position = order

    def __repr__(self):
        return str(self.value)
```

```python
listItems = []
chronoListItems = []
zeroIndex = -1
with open(inputfilepath,'r') as f:
    order = 1
    for line in f:
        value = int(line.rstrip())
        item = Item(value,order)
        if value == 0:
            zeroIndex = order
        order += 1
        listItems.append(item)
        chronoListItems.append(item)
#print(listItems)
```

```python
def findItemLocation(list,objectOrder):
    for index, object in enumerate(list):
        if object.chronological_order == objectOrder:
            return index
    else:
        print(f'Not found item {objectOrder}')
        return None
```

```python
numItems = len(listItems)
print(numItems)
copyListItems = copy.deepcopy(listItems)
print(copyListItems)
for item in range(1, numItems + 1):
    position = findItemLocation(copyListItems,item)   
    value = copyListItems[position].value
    new_position = (position + value) % (numItems-1)
    #print(f'position {position}, value {value}, new_position {new_position}')
    object = copyListItems.pop(position)
    if new_position == 0:
        copyListItems = copyListItems + [object]
    elif new_position == numItems - 1:
        copyListItems = copyListItems + [object]
    else:
        copyListItems = copyListItems[0:new_position] + [object] + copyListItems[new_position:]
print(copyListItems)
```

```python
zeroPosition = findItemLocation(copyListItems,zeroIndex)
firstPosition = (zeroPosition + 1000) % (numItems)
secondPosition = (zeroPosition + 2000) % (numItems)
thirdPosition = (zeroPosition + 3000) % (numItems)
results = [copyListItems[firstPosition].value, copyListItems[secondPosition].value, copyListItems[thirdPosition].value]
sum(results), results
```
