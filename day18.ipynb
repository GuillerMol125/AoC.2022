```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy
import logging

inputfilepath = 'C:\EMOLGUI\A_O_C\Day18.txt'
#inputfilepath = 'C:\EMOLGUI\A_O_C\Day18.sample.txt'
```

```python
dropletSet = set()
with open(inputfilepath,'r') as f:
    for line in f:
        droplet = list(line.rstrip().split(','))
        droplet= tuple([int(droplet[0]),int(droplet[1]),int(droplet[2])])
        dropletSet.add(droplet)
```

```python
sidesList = []
for droplet in dropletSet:
    sides = 6
    leftdroplet = tuple([droplet[0]-1,droplet[1],droplet[2]])
    if leftdroplet in dropletSet:
        sides -= 1
    rightdroplet = tuple([droplet[0]+1,droplet[1],droplet[2]])
    if rightdroplet in dropletSet:
        sides -= 1
    beforedroplet = tuple([droplet[0],droplet[1]-1,droplet[2]])
    if beforedroplet in dropletSet:
        sides -= 1
    behinddroplet = tuple([droplet[0],droplet[1]+1,droplet[2]])
    if behinddroplet in dropletSet:
        sides -= 1
    upperdroplet = tuple([droplet[0],droplet[1],droplet[2]-1])
    if upperdroplet in dropletSet:
        sides -= 1
    lowerdroplet = tuple([droplet[0],droplet[1],droplet[2]+1])
    if lowerdroplet in dropletSet:
        sides -= 1

    sidesList.append(sides)
```
