```python
import os
import sys
import numpy as np
from collections import defaultdict

inputfilepath = 'C:\EMOLGUI\A_O_C\Day5.txt'
```

```python
def defaultEntry():
    return []

def moveCrate(cratelist,fromCrate,toCrate):
    cratelist[toCrate] = cratelist[fromCrate][0:1] + cratelist[toCrate]
    cratelist[fromCrate].pop(0)
```

```python
with open(inputfilepath,'r') as f:
    crate_list = defaultdict(defaultEntry)
    line = f.readline().rstrip()
    while not '1' in line:
        crates = list(line)[1::4]
        print(line,crates)
        for index, crate in enumerate(crates):
            if not crate == ' ':
                crate_list[index + 1].append(crate)
        line = f.readline().rstrip()
    while not 'move' in line:
        line = f.readline().rstrip()
    totalOps = 0    
    while 'move' in line:
        totalOps += 1
        order = line.split()
        #print(order)
        numOps = int(order[1])
        fromWhere = int(order[3])
        toWhere = int(order[5])
        for opCount in range(numOps):
            moveCrate(crate_list,fromWhere,toWhere)
        #print(crate_list)
        line = f.readline().rstrip()
        
```
