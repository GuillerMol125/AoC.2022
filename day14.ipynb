```python
import os
import sys
import numpy as np
from collections import defaultdict
import gmpy2
import copy

inputfilepath = 'C:\EMOLGUI\A_O_C\Day14.txt'
```

```python
with open(inputfilepath,'r') as f:
    minHorPos = 500
    maxHorPos = 500
    maxDepth = 0
    rockPaths = []
    for line in f:
        lineContent = line.rstrip().split('->')
        curPath = []
        for lineItem in lineContent:
            pathPoint = tuple([int(pathCoord) for pathCoord in lineItem.split(',')])
            minHorPos = min(minHorPos, pathPoint[0])
            maxHorPos = max(maxHorPos, pathPoint[0])
            maxDepth = max(maxDepth, pathPoint[1])
            curPath.append(pathPoint)
        rockPaths.append(curPath)
minHorPos -= 1
maxHorPos += 1
```

```python
rockPaths
```

```python
Cave = np.zeros((maxHorPos-minHorPos + 1,maxDepth + 1))
Cave.shape, maxHorPos, minHorPos, maxDepth
```

```python
def drawRockSegment(cave, init, end, minHorPos):
    x_init, y_init = min(init[0]-minHorPos, end[0]-minHorPos), min(init[1], end[1])
    x_end, y_end = max(init[0]-minHorPos, end[0]-minHorPos), max(init[1], end[1])
    print('drawing segment',x_init,x_end,y_init,y_end)
    for x_pos in range(x_init,x_end+1):
        for y_pos in range(y_init,y_end+1):
            print('drawing rock')
            cave[x_pos,y_pos]=1
```

```python
copyRockPaths = copy.deepcopy(rockPaths)
for path in copyRockPaths:
    init = path.pop(0)
    while path:
        end = path.pop(0)
        drawRockSegment(Cave,init,end, minHorPos)
        init = end
```

```python
sandCave = copy.deepcopy(Cave)
print(f'Origin point {500 - minHorPos, 0}')
while True:
    path = []
    x_sandGrain = 500 - minHorPos
    y_sandGrain = 0
    path.append((x_sandGrain, y_sandGrain))
    while y_sandGrain < maxDepth:
        if sandCave[x_sandGrain,y_sandGrain + 1] == 0:
            y_sandGrain += 1
        elif sandCave[x_sandGrain - 1,y_sandGrain + 1] == 0:
            y_sandGrain += 1
            x_sandGrain -= 1
        elif sandCave[x_sandGrain + 1,y_sandGrain + 1] == 0:    
            y_sandGrain += 1
            x_sandGrain += 1
        else:
            sandCave[x_sandGrain,y_sandGrain] = 2
            break
        path.append((x_sandGrain, y_sandGrain))
    else:
        print(x_sandGrain, y_sandGrain)
        break
```

```python
for point in path:
    print(point,sandCave[point[0],point[1]],sandCave[point[0]-1:point[0]+2,point[1]+1])
```

```python
restingGrainsOfSand = np.where(sandCave == 2)
len(restingGrainsOfSand[0]), len(restingGrainsOfSand[1])
```
